```python
import os
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F

from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.metrics import r2_score, accuracy_score
from sklearn.model_selection import train_test_split

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

def set_seed(seed=0):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(0)
```

```python
class LatentGaussianTwoView:
    def __init__(
        self,
        K=8,
        dx=16,
        dy=16,
        sigma_x=0.35,
        sigma_y=1.00,
        mean_scale=3.0,
        imbalance=1.0,
        seed=0,
    ):
        """
        K: number of latent classes
        dx, dy: dimensions of x and y
        sigma_x: noise level of x-view. Smaller means x almost reveals class.
        sigma_y: noise level of y-view. Larger means y is more ambiguous.
        imbalance: class prior imbalance. pi_k proportional to (k+1)^(-imbalance).
                   imbalance=0 gives uniform prior.
        """
        self.K = K
        self.dx = dx
        self.dy = dy
        self.sigma_x = sigma_x
        self.sigma_y = sigma_y
        self.mean_scale = mean_scale
        self.imbalance = imbalance

        rng = np.random.default_rng(seed)

        weights = np.array([(k + 1) ** (-imbalance) for k in range(K)], dtype=np.float64)
        self.pi = weights / weights.sum()

        # Random class means, normalized and scaled.
        mux = rng.normal(size=(K, dx))
        muy = rng.normal(size=(K, dy))

        mux = mux / np.linalg.norm(mux, axis=1, keepdims=True) * mean_scale
        muy = muy / np.linalg.norm(muy, axis=1, keepdims=True) * mean_scale

        self.mux = mux.astype(np.float32)
        self.muy = muy.astype(np.float32)

    def sample(self, n, seed=None):
        rng = np.random.default_rng(seed)
        s = rng.choice(self.K, size=n, p=self.pi)

        x = self.mux[s] + self.sigma_x * rng.normal(size=(n, self.dx))
        y = self.muy[s] + self.sigma_y * rng.normal(size=(n, self.dy))

        return (
            x.astype(np.float32),
            y.astype(np.float32),
            s.astype(np.int64),
        )

    def posterior_x(self, x):
        return self._posterior_gaussian(x, self.mux, self.sigma_x)

    def posterior_y(self, y):
        return self._posterior_gaussian(y, self.muy, self.sigma_y)

    def _posterior_gaussian(self, z, means, sigma):
        """
        Compute p(k | z) exactly under Gaussian mixture with shared covariance.
        """
        z = np.asarray(z, dtype=np.float32)
        means = np.asarray(means, dtype=np.float32)

        # log pi_k - ||z - mu_k||^2 / (2 sigma^2)
        sqdist = ((z[:, None, :] - means[None, :, :]) ** 2).sum(axis=2)
        logits = np.log(self.pi[None, :] + 1e-30) - sqdist / (2.0 * sigma ** 2)

        logits = logits - logits.max(axis=1, keepdims=True)
        probs = np.exp(logits)
        probs = probs / probs.sum(axis=1, keepdims=True)
        return probs.astype(np.float32)

    def log_ratio_y(self, y, tau=1.0):
        """
        tau * log(p(k | y) / pi_k), the theory-predicted target for g(y).
        """
        py = self.posterior_y(y)
        target = tau * (np.log(py + 1e-12) - np.log(self.pi[None, :] + 1e-12))
        return target.astype(np.float32)

    def log_ratio_x(self, x, tau=1.0):
        px = self.posterior_x(x)
        target = tau * (np.log(px + 1e-12) - np.log(self.pi[None, :] + 1e-12))
        return target.astype(np.float32)
```

```python
class MLPEncoder(nn.Module):
    def __init__(self, in_dim, out_dim, hidden=128, depth=2, normalize=False):
        super().__init__()
        layers = []
        d = in_dim
        for _ in range(depth):
            layers.append(nn.Linear(d, hidden))
            layers.append(nn.ReLU())
            d = hidden
        layers.append(nn.Linear(d, out_dim))
        self.net = nn.Sequential(*layers)
        self.normalize = normalize

    def forward(self, x):
        z = self.net(x)
        if self.normalize:
            z = F.normalize(z, dim=-1)
        return z


class ContrastiveModel(nn.Module):
    def __init__(self, dx, dy, D, hidden=128, depth=2, normalize=False):
        super().__init__()
        self.f = MLPEncoder(dx, D, hidden=hidden, depth=depth, normalize=normalize)
        self.g = MLPEncoder(dy, D, hidden=hidden, depth=depth, normalize=normalize)

    def forward(self, x, y):
        fx = self.f(x)
        gy = self.g(y)
        logits = fx @ gy.T
        return logits, fx, gy


def symmetric_infonce_loss(logits, tau=1.0):
    """
    Standard CLIP-style symmetric InfoNCE.
    Positive pairs are diagonal entries.
    """
    logits = logits / tau
    n = logits.shape[0]
    labels = torch.arange(n, device=logits.device)
    loss_xy = F.cross_entropy(logits, labels)
    loss_yx = F.cross_entropy(logits.T, labels)
    return 0.5 * (loss_xy + loss_yx)


@torch.no_grad()
def retrieval_top1(model, x, y, batch_size=1024, tau=1.0):
    model.eval()
    x_t = torch.tensor(x, dtype=torch.float32, device=device)
    y_t = torch.tensor(y, dtype=torch.float32, device=device)

    fx_list = []
    gy_list = []

    for i in range(0, len(x), batch_size):
        fx_list.append(model.f(x_t[i:i+batch_size]).cpu())

    for i in range(0, len(y), batch_size):
        gy_list.append(model.g(y_t[i:i+batch_size]).cpu())

    fx = torch.cat(fx_list, dim=0)
    gy = torch.cat(gy_list, dim=0)

    scores = fx @ gy.T
    pred = scores.argmax(dim=1).numpy()
    labels = np.arange(len(x))
    return (pred == labels).mean()
```

```python
def train_contrastive(
    data,
    D=8,
    n_train=20000,
    n_val=3000,
    batch_size=512,
    epochs=20,
    lr=1e-3,
    tau=0.5,
    hidden=128,
    depth=2,
    normalize=False,
    seed=0,
):
    set_seed(seed)

    x_train, y_train, s_train = data.sample(n_train, seed=seed + 1)
    x_val, y_val, s_val = data.sample(n_val, seed=seed + 2)

    model = ContrastiveModel(
        dx=data.dx,
        dy=data.dy,
        D=D,
        hidden=hidden,
        depth=depth,
        normalize=normalize,
    ).to(device)

    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)

    x_train_t = torch.tensor(x_train, dtype=torch.float32)
    y_train_t = torch.tensor(y_train, dtype=torch.float32)

    n = len(x_train)

    for epoch in range(1, epochs + 1):
        model.train()
        perm = torch.randperm(n)
        losses = []

        for start in range(0, n, batch_size):
            idx = perm[start:start + batch_size]
            xb = x_train_t[idx].to(device)
            yb = y_train_t[idx].to(device)

            logits, _, _ = model(xb, yb)
            loss = symmetric_infonce_loss(logits, tau=tau)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            losses.append(loss.item())

        if epoch == 1 or epoch % max(1, epochs // 5) == 0:
            top1 = retrieval_top1(model, x_val[:1000], y_val[:1000], tau=tau)
            print(
                f"D={D:>3d} | epoch {epoch:>3d}/{epochs} | "
                f"loss={np.mean(losses):.4f} | val top1={top1:.4f}"
            )

    return model, (x_train, y_train, s_train), (x_val, y_val, s_val)
```

```python
@torch.no_grad()
def get_embeddings(model, x, y, batch_size=1024):
    model.eval()

    x_t = torch.tensor(x, dtype=torch.float32, device=device)
    y_t = torch.tensor(y, dtype=torch.float32, device=device)

    fx_list = []
    gy_list = []

    for i in range(0, len(x), batch_size):
        fx_list.append(model.f(x_t[i:i+batch_size]).cpu().numpy())

    for i in range(0, len(y), batch_size):
        gy_list.append(model.g(y_t[i:i+batch_size]).cpu().numpy())

    return np.concatenate(fx_list, axis=0), np.concatenate(gy_list, axis=0)


def ridge_r2(source, target, alpha=1e-3):
    """
    Fit linear map source -> target and report test R^2.
    """
    X_train, X_test, Y_train, Y_test = train_test_split(
        source, target, test_size=0.35, random_state=0
    )

    reg = Ridge(alpha=alpha)
    reg.fit(X_train, Y_train)
    pred = reg.predict(X_test)

    return r2_score(Y_test, pred, multioutput="variance_weighted")


def linear_probe_acc(source, labels):
    """
    Linear classifier accuracy from embedding to latent class.
    """
    X_train, X_test, y_train, y_test = train_test_split(
        source, labels, test_size=0.35, random_state=0, stratify=labels
    )

    clf = LogisticRegression(
        max_iter=2000,
        C=1.0,
        multi_class="auto",
    )
    clf.fit(X_train, y_train)
    pred = clf.predict(X_test)
    return accuracy_score(y_test, pred)


def evaluate_model(data, model, val_data, tau=0.5):
    x_val, y_val, s_val = val_data
    fx, gy = get_embeddings(model, x_val, y_val)

    posterior_x = data.posterior_x(x_val)
    posterior_y = data.posterior_y(y_val)
    log_ratio_y = data.log_ratio_y(y_val, tau=tau)
    log_ratio_x = data.log_ratio_x(x_val, tau=tau)

    metrics = {}

    # Theory-predicted alignments
    metrics["r2_f_to_posterior_x"] = ridge_r2(fx, posterior_x)
    metrics["r2_g_to_log_ratio_y"] = ridge_r2(gy, log_ratio_y)

    # Alternative targets, useful for comparison
    metrics["r2_f_to_log_ratio_x"] = ridge_r2(fx, log_ratio_x)
    metrics["r2_g_to_posterior_y"] = ridge_r2(gy, posterior_y)

    # Linear class probes
    metrics["class_acc_from_f"] = linear_probe_acc(fx, s_val)
    metrics["class_acc_from_g"] = linear_probe_acc(gy, s_val)

    # Pair retrieval
    metrics["retrieval_top1"] = retrieval_top1(model, x_val[:1500], y_val[:1500], tau=tau)

    # Score spectrum
    fx_small = fx[:1000]
    gy_small = gy[:1000]
    score = fx_small @ gy_small.T
    singular_values = np.linalg.svd(score, compute_uv=False)
    metrics["top_singular_values"] = singular_values

    return metrics
```
